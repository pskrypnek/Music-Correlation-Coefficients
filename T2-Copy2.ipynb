```python
ccclist = []

with open('Country_Correlation_Coefficients.csv', 'w') as file:
    writer = csv.writer(file, delimiter = ';')
    writer.writerow(np.linspace(1,50,50))
    for a in range(49):
        List = []
        for b in range(49):
            c = np.corrcoef(compare[a], compare[b])[1,0]
            List.append(c)
        s = sum(List)/49
        List.append(s)
        ccclist.append(s)
        writer.writerow(List)  
```

```python
ccrlist = []
with open('R&B_Correlation_Coefficients.csv', 'w') as file:
    writer = csv.writer(file, delimiter = ';')
    writer.writerow(np.linspace(1,50,50))    
    for d in range(50,99):
        List1 = []
        for e in range(50,99):
            c = np.corrcoef(compare[d], compare[e])[1,0]
            List1.append(c)
        s = sum(List1)/49
        List1.append(s)
        ccrlist.append(s)
        writer.writerow(List1)
```

```python
print("Country Music Correlation Coefficient" + " " + str(sum(ccclist)/49))
```

```python
print("R&B Music Correlation Coeficient" + " " + str(sum(ccrlist)/49))
```
